```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import GridSearchCV
from numpy import mean
from sklearn.metrics import f1_score
import statistics as st
```

```python
dataset = pd.read_csv('Modified(1).csv')
dataset
```

```python
input = dataset.drop(['Output'], axis=1 )
X = np.array(input)
X
```

```python
output = dataset['Output']
y = np.array(output)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.1, train_size = 0.9, random_state = 0)
len(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler(copy = True)

sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.99)
```

```python
pca.fit(X_train)
pca.n_components_
```

```python
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 3, shuffle = True, random_state = 0)
kf
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
params = {'n_estimators':[1000, 1500, 1600], 'max_depth':[1, 5, 10], 'random_state': [0, 1]}
rfcp = GridSearchCV(rfc,params)
```

```python
for train_index, test_index in kf.split(X):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    rfcp.fit(X_train, y_train)
    y_predr = rfcp.predict(X_test)
    acc = rfcp.score(X_test, y_test)
    f1r = f1_score(y_test, y_predr, average = 'micro')
    print(acc)
    print('\n')
print(rfcp.best_params_)
acg = mean(acc)
f1ravg = mean(f1r)
print('\n The average accuracy:', acc)
print('\n The average F1 score:', f1ravg)
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC()
paramss = {'C':[0.01, 0.1, 0.5, 1.0, 1.5],'dual':[True],'random_state':[0, 1, 2],  'max_iter':[1000, 1500]}
svcp = GridSearchCV(svc, paramss)
```

```python
for train_index, test_index in kf.split(X):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    svcp.fit(X_train, y_train)
    y_preds = svcp.predict(X_test)
    accs = svcp.score(X_test, y_test)
    f1s = f1_score(y_test, y_preds, average = 'micro')
    print(accs)
    print('\n')
print(svcp.best_params_)
acgs = mean(accs)
f1savg = mean(f1s)
print('\n The average accuracy:', accs)
print('\n The average F1 score:', f1savg)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
paramsss = {'n_neighbors':[1, 3, 5]}
knnp = GridSearchCV(knn, paramsss)
```

```python
for train_index, test_index in kf.split(X):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    knnp.fit(X_train, y_train)
    y_predk = knnp.predict(X_test)
    acck = knnp.score(X_test, y_test)
    f1k = f1_score(y_test, y_predk, average = 'micro')
    print(acck)
    print('\n')
print(knnp.best_params_)
acgk = mean(acck)
f1kavg = mean(f1k)
print('\n The average accuracy:', acck)
print('\n The average F1 score:', f1kavg)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
acc = classifier.score(X_test, y_test)
acc
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Predicting the training set
# result through scatter plot
from matplotlib.colors import ListedColormap
 
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1,
                     stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1,
                     stop = X_set[:, 1].max() + 1, step = 0.01))
 
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(),
             X2.ravel()]).T).reshape(X1.shape), alpha = 0.75,
             cmap = ListedColormap(('yellow', 'white', 'aquamarine')))
 
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
 
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
 
plt.title('Logistic Regression (Training set)')
plt.xlabel('PC1') # for Xlabel
plt.ylabel('PC2') # for Ylabel
plt.legend() # to show legend
 
# show scatter plot
plt.show()
```
